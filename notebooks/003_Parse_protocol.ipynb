```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/protocols/protocol_trace.csv", usecols=["t", "v"])
```

```python
plt.plot(df.t, df.v, '-')
```

```python
mask_changes = df.diff().v != 0
mask_changes.iloc[-1] = True
df = df[mask_changes]

plt.plot(df.t, df.v, '-')
```

```python
df.to_csv("../data/protocols/protocol_sparse.csv", index=False)
```
